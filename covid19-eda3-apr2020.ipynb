```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import os
import wget
```

```python
data_file = "data-dssg-covid19pt.csv"

if (os.path.exists(data_file)):
    os.remove(data_file)

url = f"https://raw.githubusercontent.com/dssg-pt/covid19pt-data/master/data.csv"

wget.download(url, data_file)
```

```python
covdf = pd.read_csv('data-dssg-covid19pt.csv')
```

```python
covdf.tail()
```

```python
covdf1 = covdf.drop([
    'confirmados_estrangeiro',
    'cadeias_transmissao',
    'obitos_estrangeiro',
    'recuperados_arsnorte',
    'recuperados_arscentro',
    'recuperados_arslvt',
    'recuperados_arsalentejo',
    'recuperados_arsalgarve',
    'recuperados_acores',
    'recuperados_madeira',
    'recuperados_estrangeiro'
    ], axis=1)
```

```python
covdf1.head()
```

## Find initial row and value on each temporal sequence

```python
nCols = covdf1.shape[1]

# lists to store the index,value where valid data (not NaN) start, for each column
row_start_index = []
row_start_value = []

row_start_index = [0]*nCols       # set all indexes to zero
row_start_value = [np.nan]*nCols  # set all values to NaN

for row_tuple in covdf1.itertuples():
    cols = list(row_tuple)
    for nc in range(3,nCols):
        if np.isnan(cols[nc])==False and np.isnan(row_start_value[nc])==True:
            row_start_index[nc] = cols[0]
            row_start_value[nc] = cols[nc]
```

```python
print(f"Initial sequence's index: {row_start_index}")
print(f"Initial sequence's value: {row_start_value}")
```

## Fill NaN cells with zero

```python
# Fill NaN with ZERO

covdf2 = covdf1.fillna(0,inplace=False)

covdf2.head()

# Alternative: replace NaN by zero in place
# covdf1.fillna(0,inplace=True)
# covdf1.head()
```

```python
covdf2.describe().transpose()
```

### Compute difference among sucessive accumlated values to get daily statistics

```python
colNames = covdf2.columns

# print(colNames)
```

```python
# exclude first 2 columns ('data', 'data_dados') from computing daily values

for nc in range(2,len(colNames)):
    src_name = colNames[nc]
    dst_name = src_name+"_dia"
    # print(dst_name)
    covdf2[dst_name] = covdf2[src_name].diff(1)
    covdf2.at[0,dst_name] = covdf2.at[0,src_name] 
```
